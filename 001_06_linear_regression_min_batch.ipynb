# B. 다중 선형 회기 구현하기

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

torch.manual_seed(1)
```

1. data

```python
# 데이터
x_train = torch.FloatTensor([[73, 80, 75],
                             [93, 88, 93],
                             [89, 91, 90],
                             [96, 98, 100],
                             [73, 66, 70]])
y_train = torch.FloatTensor([[152], [185], [180], [196], [142]])
```

2. class

```python
class MulitvaariateLinearRegressionModel(nn.Module):
  def __init__(self):
    super().__init__()
    self.linear = nn.Linear(3,1)

  def forward(self, x):
    return self.linear(x)
```

3. Model

```python
model = MulitvaariateLinearRegressionModel()
```

4. optimizer

```python
optimizer = torch.optim.SGD(model.parameters(), lr=1e-5)
```

5. training


```python
epoches = 2000

for epoch in range(epoches):
  # Hx
  prediction = model(x_train)

  # cost
  cost = F.mse_loss(prediction, y_train)

  # 
  optimizer.zero_grad()
  cost.backward()
  optimizer.step()

  if epoch % 100 == 0:
    # 100번마다 로그 출력
      print('Epoch {:4d}/{} Cost: {:.6f}'.format(
          epoch, epoces, cost.item()
      ))
  
```

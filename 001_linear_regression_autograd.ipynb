자동 미분(Autograd)
 requires_grad=True, backward()

자동 미분(Autograd) 실습하기

```python
import torch
```

1. 모델: H(x) 2x^2 + 5 

```python
# 값이 2인 스칼라 텐서 w
# required_grad를 True로 설정
w = torch.tensor(2.0, requires_grad=True)
# w에 텐서에 대한 기울기가 저장

y = w**2
z = 2*y + 5
```

 2. 해당 수식을 w에 대해서 미분
 .backward()를 호출하면 해당 수식의 w에 대한 기울기를 계산

```python
z.backward()

print('수식을 w로 미분한 값 : {}'.format(w.grad))
```
